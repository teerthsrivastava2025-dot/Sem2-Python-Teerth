```python
#DATING APP

def create_profile(name, age, interest, city):
    profile = {
    'name': name,
    'age': age,
    'interest': interest,
    'city': city
    }
    return profile
```

```python
create_profile('Ashish', 24, 'Music', 'Pune')
```

```python
def calculate_match_score(profile1, profile2):
    score = 0
    
    #same interest
    if profile1['interest'] == profile2['interest']:
        score = score + 50
    
    #same city
    if profile1['city'] == profile2['city']:
        score = score + 30
    
    age_diff = abs(profile1['age'] - profile2['age'])
    if age_diff <= 2 :
        score = score + 20
    
    return score 
```

```python
def is_match(score):
    if score >= 60:
        return True
    else:
        return False
```

```python
is_match(50)
```

```python
is_match(90)
```

```python
def display_results(profile1, profile2, score):
    print('Welcome to the Dating App')
    print('=' * 40)
    print('Person 1:', profile1['name'])
    print('Person 2:', profile2['name'])
    print('Compatibility Score:', score)
    
    if is_match(score):
        print('Its a match!!')
    else:
        print('Sorry  better luck next time')
```

```python
# RUN THE PROGRAM

user1 = create_profile('Ashish', 21, 'Music', 'Delhi')
user2 = create_profile('Ashika', 19, 'Trekking', 'Pune')

match_score = calculate_match_score(user1, user2)

display_results(user1, user2, match_score)
```
