```python
#ADVENTURE GAME
def start_game():
    print('Welcome to treasure game')
    print("You have hit a crossroad")
    print("Forest or Village")
    
    choice=input('Where will you go:')
    if choice=='Forest':
        forest()
    elif choice=='Village':
        print('Lets go to the Village')
    else:
        print('Invalid Output')
        
#FOREST
def forest():
    print('Welcome to the forest')
    print("You have hit a crossroad")
    print("Cave or River")
    
    choice=input('Where will you go:')
    if choice=='Cave':
        cave()
    elif choice=='River':
        print('You drowned in the River')
    else:
        print('Invalid Output')
        
#Cave
def cave():
    print('Welcome to the Cave')
    print("You have hit a crossroad")
    print("Wise old elf or Ignore the elf")
    
    choice=input('Where will you go:')
    if choice=='Wise old elf':
        print('The wise old elf gives you the treasure! üèÜ You win!')
    elif choice=='Ignore the elf':
        print('You are lost in the mystic caves forever')
    else:
        print('Invalid Output')
    
start_game()
```
